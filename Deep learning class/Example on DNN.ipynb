# Regression 

```python
X, y = datasets.make_regression(n_samples= 1000, n_features= 2, noise= 9)

# creating a dataframe to split the data into two vectors
# each for one class
df = pd.DataFrame(dict(x1 = X[:,0], x2 = X[:,1], label = y))

# scatter plot to see the clusters
sns.scatterplot(x = "x1", y = "x2", data = df)

# Normalize the data using Standard scalar
scalar = StandardScaler()
scalar.fit(X)
X = scalar.transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## model10(with "gd")

```python
model10 = DNN( [20, 10, 5, 1], ["relu", "relu", "relu", "linear"],learning_rate= 0.01, max_epochs = 5000, learning_algo = "gd", model_type= "reg", cost_reg= "RMSE")
cost, epochs = model10.fit(X_train, y_train)
```

### cost vs epochs

```python
plt.plot(epochs, cost)
plt.xlabel("epochs")
plt.ylabel('cost')
```

```python
y_hat = model10.predict(X_train)
y_pred = model10.predict(X_test)
```

```python
# evaluating training

# mean absolute error
mae = mean_absolute_error(y_train, y_hat)
print("mean absolute error for training = ", mae)

# mean squared error
mse = mean_squared_error(y_train, y_hat)
print("mean squared error for training = ", mse)

# R2 score (value = 1, means fitting perfectly, value = 0, very poor learning)
r2 = r2_score(y_train, y_hat)
print("R2 score for training = ", r2)
```

```python
# evaluating training

# mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("mean absolute error for testing = ", mae)

# mean squared error
mse = mean_squared_error(y_test, y_pred)
print("mean squared error for testing = ", mse)

# R2 score (value = 1, means fitting perfectly, value = 0, very poor learning)
r2 = r2_score(y_test, y_pred)
print("R2 score for testing = ", r2)
```
